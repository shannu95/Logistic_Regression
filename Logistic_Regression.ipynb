```python
# Importing the libraries
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
%matplotlib inline
```

```python
# Importing the dataset
dataset = pd.read_csv('bank_details.csv')
dataset.head()
```

```python
dataset.isnull().sum()
```

```python
sns.countplot(x='term_deposit',data=dataset,hue='marital')
```

```python
sns.distplot(dataset['age'],kde=False)
```

```python
dataset.info()
```

```python
job_c=pd.get_dummies(dataset['job'],drop_first=True,prefix='job')
marital_c=pd.get_dummies(dataset['marital'],drop_first=True,prefix='marital')
education_c=pd.get_dummies(dataset['education'],drop_first=True,prefix='education')
default_c=pd.get_dummies(dataset['default'],drop_first=True,prefix='default')
housing_c=pd.get_dummies(dataset['housing'],drop_first=True,prefix='housing')
loan_c=pd.get_dummies(dataset['loan'],drop_first=True,prefix='loan')
contact_c=pd.get_dummies(dataset['contact'],drop_first=True,prefix='contact')
month_c=pd.get_dummies(dataset['month'],drop_first=True,prefix='month')
day_of_week_c=pd.get_dummies(dataset['day_of_week'],drop_first=True,prefix='day_of_week')
poutcome_c=pd.get_dummies(dataset['poutcome'],drop_first=True,prefix='poutcome')
dataset=pd.concat([dataset,job_c,marital_c,education_c,default_c,housing_c,loan_c,contact_c,month_c,day_of_week_c,poutcome_c],axis=1)
```

```python
dataset.head()
```

```python
dataset.drop(['job','marital','education','default','housing','loan','contact','month','day_of_week','poutcome'],axis=1,inplace=True)
```

```python
dataset.head()
```

```python
cols = list(dataset.columns.values)
cols
```

```python
X = dataset.drop('term_deposit',axis=1)
Y = dataset['term_deposit']
```

```python
X.head()
```

```python
Y.head()
```

```python
# Splitting the dataset into the Training set and Test set
from sklearn.cross_validation import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size = 0.25, random_state = 0)
```

```python
# Fitting Logistic Regression to the Training set
from sklearn.linear_model import LogisticRegression
classifier = LogisticRegression()
classifier.fit(X_train, y_train)
```

```python
# Predicting the Test set results
y_pred = classifier.predict(X_test)
```

```python
from sklearn.metrics import classification_report
print(classification_report(y_test,y_pred))
```

```python
# Making the Confusion Matrix
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(y_test, y_pred)
```
